### **Brute Force Solution**

```python
def maxProfit(arr):

    n = len(arr)
    max_profit = 0

    for i in range(n):
        for j in range(1+1, n):
            profit = arr[j]-arr[i]
            max_profit = max(max_profit, profit)

    return max_profit

arr = [7,1,5,3,6,4]
print(maxProfit(arr))


# Time Complexity => O(n²)  - Two nested loops.
# Space Complexity => O(1)
```

### **Optimal Solution (Track Minimum Price)**

```python
def maxProfit(arr):
    
    min_price = arr[0]
    max_profit = 0

    for i in range(len(arr)):

        profit = arr[i] - min_price
        max_profit = max(max_profit, profit)

        min_price = min(min_price, arr[i])

    return max_profit

arr = [7,1,5,3,6,4]
print(maxProfit(arr))


# Time Complexity => O(n)  - Single Traversal
# Space Complexity => O(1)
```
