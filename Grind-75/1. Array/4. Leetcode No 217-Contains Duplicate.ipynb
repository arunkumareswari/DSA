### **Brute Force Solution**

```python
def containsDuplicate(arr):

    n = len(arr)
    for i in range(n):
        for j in range(i+1, n):
            if arr[i] == arr[j]:
                return True
    return False

arr = [1, 2, 3, 4, 3]
print(containsDuplicate(arr))


# Time  : O(N²)
# Space : O(1)
```

### **Better Solution (Sorting)**

```python
def containsDuplicate(arr):

    arr.sort()

    for i in range(len(arr)):
        if arr[i] == arr[i+1]:
            return True
    return False


arr = [1, 2, 3, 4, 3]
print(containsDuplicate(arr))


# Time  : O(N log N)
# Space : O(1)
```

### **Optimal Solution (HashSet)**

```python
def containsDuplicate(arr):

    hashset = {}

    for num in arr:
        if num in hashset:
            return True
        hashset[num] = 1

    return False 

arr = [1, 2, 4, 3]
print(containsDuplicate(arr))


# Time  : O(N)
# Space : O(N)
```
