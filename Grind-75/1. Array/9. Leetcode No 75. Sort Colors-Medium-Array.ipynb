
```
You are given an array with only:

0 → Red
1 → White
2 → Blue

You must sort it in-place.

Example:

Input  = [2,0,2,1,1,0]
Output = [0,0,1,1,2,2]
```

---

### **Brute Force (Sorting)**

````python
def sortColor(arr):
    arr.sort()
    return arr

arr = [2,0,2,1,1,0]
print(sortColor(arr))

# Time Complexity: O(n log n)
# Space Complexity: O(1)
````

### **Better Approach (Counting Sort)**

````python
# Idea : Count number of 0s, 1s, 2s and overwrite array.

def sortColor(arr):

    count0 = count1 = count2 = 0
    for num in arr:
        if num == 0:
            count0 += 1
        elif num == 1:
            count1 += 1
        else: 
            count2 += 1

    i = 0

    for _ in range(count0): #  "_" --> dummy variable or throwaway vaiable
        arr[i] = 0
        i += 1

    for _ in range(count1):
        arr[i] = 1
        i += 1

    for _ in range(count2):
        arr[i] = 2
        i += 1

    return arr

arr = [2,0,2,1,1,0]
print(sortColor(arr))

# Time Complexity: O(n)   <=  O(n) + O(n) = O(n)
# Space Complexity: O(1)
````

### **Optimal Approach (Dutch National Flag Algorithm)**

````python
# Idea

# Use 3 pointers:

# low → place 0
# mid → current element
# high → place 2
````

````python
def sortColor(arr):

    low = 0
    mid = 0
    high = len(arr) - 1

    while mid < high:

        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1
            mid += 1

        elif arr[mid] == 1:
            mid += 1
        
        else:
            arr[mid], arr[high] = arr[high], arr[mid]
            high -= 1

    return arr

arr = [2,0,2,1,1,0]
print(sortColor(arr))

# Time Complexity: O(n)
# Space Complexity: O(1)
````
