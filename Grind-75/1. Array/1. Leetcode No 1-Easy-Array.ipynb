### **Brute Force**

```python
def two_sum(arr, target):

    n = len(arr)
    for i in range(n):

        for j in range(1+i,n):
            if arr[i]+arr[j] == target:
                return [i,j]
            
    return []


arr = [2, 11, 15, 7]
target = 9
result = two_sum(arr, target)
print(result)

# Time Complexity => O(n²)
# Space Complexity => O(1)
```

### **Better Solution (Sorting + Two Pointers)**

```python
def two_sum(arr, target):

    arr = [ (arr[i],i) for i in range(len(arr))] # output : [(2, 0), (11, 1), (15, 2), (7, 3)]
    arr.sort()                                   # outpuy : [(2, 0), (7, 3), (11, 1), (15, 2)]
    
    left = 0
    right = len(arr) - 1

    while left < right:
        current_sum = arr[left][0] + arr[right][0]  # values: 2 + 15
        
        if current_sum == target:
            return [arr[left][1], arr[right][1]]        # index --> [1]

        elif current_sum < target:
            left +=1
        else: 
            right -= 1
    return []


arr = [2, 11, 15, 7]
target = 9
result = two_sum(arr, target)
print(result)

# Time Complexity
# Sorting: O(n log n)
# Two-pointer traversal: O(n)

# Total: O(n log n)


# Space Complexity -> O(n)
```

### **Optimal Solution (Hash Map)**

```python
# for each element : complement = target - nums[i]

def two_sum(arr, target):

    hashmap = {}

    for i in range(len(arr)):
        complement = target - arr[i]

        if complement in hashmap:
            return [hashmap[complement], i]
        
        hashmap[arr[i]] = i
    return []


arr = [2, 11, 15, 7]
target = 9
result = two_sum(arr, target)
print(result)

# Time Complexity ->  O(n)
# Space Complexity -> O(n)
```
