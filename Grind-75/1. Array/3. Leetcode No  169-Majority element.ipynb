### **Brute Force Solution**

```python

# Idea :
# For each element , count how many time it appears in the array.
# If the count is greater than n/2, return that element

def majority_element(arr):

    for i in range(len(arr)):
        count = 1
        for j in range(i+1, len(arr)):
            if arr[i] == arr[j]:
                count +=1
        
        if count > len(arr)//2:
            return arr[i]
    
    return -1


arr = [2,2,1,1,1,2,2]
result = majority_element(arr)
print(result)


# Time  : O(N²)
# Space : O(1)
```

### **Better Solution (Using Hash Map / Dictionary)**

```python

def majority_element(arr):

    freq = {}
    count = 0

    n = len(arr)
    for num in arr:

        if num in freq:
            freq[num] +=1
        else:
            freq[num] = 1

        if freq[num] < n//2:
            return num
        
    return -1



arr = [2,2,1,1,1,2,2]
result = majority_element(arr)
print(result)



# Time  : O(N)
# Space : O(N)
```

### **Optimal Solution (Moore’s Voting Algorithm)**

```python
def majority_element(arr):

    vote = 0 
    majority = None

    for num in arr:

        if vote == 0:
            majority = num

        if majority == num:
            vote +=1
        else:
            vote -=1

    return majority


arr = [2,2,1,1,1,2,2]
result = majority_element(arr)
print(result)



# Time  : O(N)
# Space : O(1)
```
