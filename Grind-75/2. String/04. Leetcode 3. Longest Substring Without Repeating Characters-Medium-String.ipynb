## **Problem: Longest Substring Without Repeating Characters**

Given a string s, find the length of the longest substring without repeating characters.

Example:

s = "abcabcbb" → Output = 3 ("abc")

--- 
### **Brute Force Approach**

>Idea:
>
>>Generate all substrings
>>
>>Check each substring for unique characters
>>
>>Track the maximum length

>Steps:
>>
>>Use two loops → generate all substrings
>>
>>For each substring:
>>
>> - Use a set to check duplicates
>>
>>Update max length

```python
def lengthOfLongestSubstring(s):

    n = len(s)
    max_length = 0

    for i in range(n):
        seen = set()

        for j in range(i, n):
            if s[j] in seen:
                break
            seen.add(s[j])
            max_length = max(max_length, j - i + 1)

    return max_length

s = "abcabcbb"
print(lengthOfLongestSubstring(s))


# Time Complexity:

# - Outer loop: O(n)
# - Inner loop: O(n)
# - Checking set: O(1)
# - Overall: O(n²)

# Space Complexity:
# - Set stores characters → O(n)
```

### **Better Approach (Sliding Window with Set)**
>Idea:
>
>>Use two pointers (left, right)
>>
>>Expand window
>>
>>If duplicate found → shrink window

>Steps:
>
>- Use a set
>
>- Move right pointer
>
>- If duplicate:
>
>   - Remove from left until unique
>
>- Update max length

```python
def lengthOfLongestSubstring(s):

    n = len(s)

    char_set = set()
    left = 0
    max_length = 0

    for right in range(n):
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1

        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)

    return max_length

s = "abcabcbb"
print(lengthOfLongestSubstring(s))


# Time Complexity: O(n)   Each character visited at most twice

# Space Complexity: O(n)
```

### **Optimal Approach (Sliding Window + HashMap)**

>Idea:
>- nstead of removing one-by-one,
>
>- Jump left pointer directly using last seen index

>Steps:
>
>- Use dictionary → char → last index
>
>- If duplicate found:
>
>- Move left = max(left, last_index + 1)
>
>- Update max length

```python
def lengthOfLongestSubstring(s):

    hash_set = {}

    left = 0
    max_length = 0

    for right in range(len(s)):
        if s[right] in hash_set:
            left = max(left, hash_set[s[right]]+1)

        hash_set[s[right]] = right
        max_length = max(max_length, right - left + 1)

    return max_length

s = "abcabcbb"
print(lengthOfLongestSubstring(s))


# Time Complexity: O(n) (single pass)

# Space Complexity: O(n)
```
