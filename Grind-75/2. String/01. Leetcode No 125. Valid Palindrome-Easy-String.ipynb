## **Problem: Valid Palindrome**

Given a string s
Ignore:

- spaces

- punctuation

- case

Check if it is a palindrome

### **Brute force Solution**

```python
# Idea:

# 1. Remove unwanted characters
# 2. Convert to lowercase
# 3. Reverse string
# 4. Compare


def isPalindrome(s):

    cleaned = ""

    for ch in s:
        if ch.isalnum():
            cleaned += ch.lower()

    return cleaned == cleaned[::-1]

s = "A man, a plan, a canal: Panama"

print(isPalindrome(s))


# Complexity:

# Time: O(n²) (because string concatenation is costly)

# Space: O(n)


# Problem:

# String is immutable in Python 

# Every time → new string created

# So internally:

# "" → "a" → "am" → "ama" → "aman" ...

# Each step copies whole string → slow
```

### **Better Approach**

```python
# Idea:
# Use list instead of string (avoids costly concatenation)
# Then join + reverse

def isPalindrome(s):

    cleaned = []

    for ch in s:
        if ch.isalnum():
            cleaned.append(ch.lower())

    cleaned_str = "".join(cleaned)

    return cleaned_str == cleaned_str[::-1]

s = "A man, a plan, a canal: Panama"

print(isPalindrome(s))

# Complexity:

# Time: O(n) 

# Space: O(n)
```

### **Optimal Approach (Two Pointer)**

```python
# Core Idea:

# 1. Use two pointers:
#     left → start
#     right → end
# 2. Skip non-alphanumeric
# 3. Compare characters directly
# 4. No extra space needed!


def isPalindrome(s):

    left = 0
    right = len(s) -1

    while left < right:
        while not isAlNum(s[left]) and left < right:
            left += 1
        while not isAlNum(s[right]) and left < right:
            right -= 1

        if s[left].upper() != s[right].upper():
            return False
        
        left += 1
        right -= 1

    return True


def isAlNum(ch):
    return ('0' <= ch <= '9') or ('A' <= ch <= 'Z') or ('a' <= ch <= 'z')


s = "A man, a plan, a canal: Panama"

print(isPalindrome(s))


# Complexity:

# Time: O(n)

# Space: O(1) 
```
