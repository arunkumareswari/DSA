>## **Problem: Valid Anagram**
>
>- Given two strings s and t
>- Check if t is an anagram of s
>- Anagram = same characters, same frequency, different order allowed
>  

### **Brute Force**

```python
# Idea:

# Sort both strings
# Compare

def isAnagram(s,t):
    return sorted(s) == sorted(t)

s = "anagram"
t = "nagaram"
print(isAnagram(s,t))

    
# Complexity

# Time: O(n log n) -> (sorting)

# Space: O(n)
```


### **Better (Using Dictionary / HashMap)**

>Idea:
>>Count frequency of each character in s
>>
>>Subtract using t
>>
>>If all counts = 0 → anagram

```python
def isAnagram(s,t):

    if len(s) != len(t):  # len - time coplexity => O(1)
        return False
    dict = {}

    for char in s:
        if char in dict:
            dict[char] += 1
        else:
            dict[char] = 1

    for char in t:
        if char in dict:
            if dict[char] != 0:
                dict[char] -= 1 
            else:
                return False
        else:
            return False

    for val in dict.values():
        if val != 0:
            return False

    return True


s = "anagram"
t = "nagaram"
print(isAnagram(s,t))

# Complexity

# Time: O(n)

# Space: O(n)
```

### **Optimal (Using Fixed Array)**
>Idea:
>>Since only lowercase letters (a-z)
>>
>>Use array of size 26
>>
>>Map char → index

```python
def isAnagram(s,t):

    if len(s) != len(t):
        return False

    count = [0]*26

    for ch in s:
        count[ord(ch) - ord('a')] += 1
    
    for ch in t:
        count[ord(ch) - ord('a')] -= 1

    return all(i == 0 for i in count)

s = "anagram"
t = "nagaram"
print(isAnagram(s,t))


# Complexity
# Time: O(n) 

# Space: O(1) 
```
