### Truth Table

| Condition A | Condition B | A ^ B |
|------------|------------|-------|
| False      | False      | False |
| False      | True       | True  |
| True       | False      | True  |
| True       | True       | False |

```python
def checkStatus(a, b, flag):
    if not flag:
        # Either a or b (not both) is non-negative
        return (a >= 0) ^ (b >= 0)
    else:
        # Both a and b are negative
        return a < 0 and b < 0

# checkStatus(1, -1, False)
checkStatus(-187, -9332, True)
# checkStatus(5, 3, True)
```

```python
def pos(n):
    for i in range(n -1, -1, -1):
        print(i, end = " ")

def neg(n):
    for i in range(n, 1):
        print(i, end= " ")



n = int(input())

if n == 0:
    print("already zero")
elif n > 0:
    pos(n)
else:
    neg(n)
```

```python
s = "Hello"
rev = ""

for i in s:
    rev = i + rev

print(rev)
```

```python
s= "Hello"  # total character = 5
rev = ""

for ch in range(len(s)-1, -1, -1):  # 5 - 1  = 4 (0 to 4)
    rev += s[ch] 
print(rev)
```

```python
word = "Hello"
reverse = ""

for i in range(len(word)-1, -1, -1):
    reverse += word[i]

print(reverse)

print(word[::-1])
```

```python
import re

def numberMatcher(str):

    pat = r"\d+"

    match = re.findall(pat, str)

    if (match):
        for i in match:
            print(i, end = " ")
    else:
        print(-1)

str = 'sdasd123asmdasdk34234kfdsd34sdfk5'

numberMatcher(str)
```
