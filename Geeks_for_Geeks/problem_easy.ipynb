```python
def frineds_in_trouble(j_angry, s_angry):

    if (j_angry and s_angry) or (not j_angry and not s_angry):   # True and True = True , False and False = True
        return True
    else:
        return False
    
print(frineds_in_trouble(True,True))
print(frineds_in_trouble(False,False))
print(frineds_in_trouble(True,False))
print(frineds_in_trouble(False, True))
    
```

```python
def frineds_in_trouble(j_angry, s_angry):
    return j_angry == s_angry                   # True == True (or) False == False -> True, so return True , True == False --> False so return False

print(frineds_in_trouble(True,True))
print(frineds_in_trouble(False,False))
print(frineds_in_trouble(True,False))
print(frineds_in_trouble(False, True))
```

```python
def fizBuzz(number):

    if (number % 3 == 0) and (number % 5 == 0):
        print("FizzBuzz")
    elif number % 3 == 0:
        print("Fizz")
    elif number % 5 == 0:
        print("Buzz")
    else:
        print(number)


fizBuzz(15)
fizBuzz(5)
fizBuzz(3)
print(fizBuzz(17))
```
